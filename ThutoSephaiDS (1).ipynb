## 10. NBA player statistics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("NBA_2024_cleaned.csv")
```

```python
df.info()
```

```python
high_scorers = df[df['PTS'] >= 20] # 'high scorers' as players with PTS >= 20
```

```python
print(high_scorers.head()) # the high scorers
```

```python
# Calculating the mean for the specified metrics for high scorers and non-high scorers
metrics = ['FGA', '3PA', 'FT', 'AST', 'TRB']
```

```python
# Calculating mean values for high scorers
high_scorers_means = high_scorers[metrics].mean()
```

```python
# Define 'non-high scorers' as players with PTS < 20
non_high_scorers = df[df['PTS'] < 20]
```

```python
# Calculating mean values for non-high scorers
non_high_scorers_means = non_high_scorers[metrics].mean()
```

```python
# Combining the results into a single DataFrame for easy comparison
comparison = pd.DataFrame({
    'High Scorers': high_scorers_means,
    'Non-High Scorers': non_high_scorers_means
})
```

```python
print(comparison) # the comparison
```

```python
# Plotting
comparison.plot(kind='bar', figsize=(10, 6))
```

```python
plt.title('Comparison of Metrics Between High Scorers and Non-High Scorers')
plt.xlabel('Metrics')
plt.ylabel('Mean Values')
```

```python
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
plt.savefig('nba_comparison_plot.png')
```
